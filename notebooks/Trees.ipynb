```python
### My solution

class BinaryTree(object):
    def __init__(self, root):
        self.root = Node(root)

    def search(self, find_val):
        """Return True if the value
        is in the tree, return
        False otherwise."""
        return self.preorder_search(self.root,find_val)

    def print_tree(self):
        """Print out all tree nodes
        as they are visited in
        a pre-order traversal."""
        traversal=""
        return self.preorder_print(self.root,traversal)
        

    def preorder_search(self, start, find_val):
        """Helper method - use this to create a 
        recursive search solution."""
        #if start.value == None:
        #    return False
        if find_val == start.value:
            return True
        
        else:
            nodeleft = start.left
            noderight = start.right
            if nodeleft is not None:
                if self.preorder_search(nodeleft,find_val):
                    return True
                else:
                    return self.preorder_search(noderight,find_val)
            elif noderight is not None:
                return self.preorder_search(noderight,find_val)
            else:
                return False
            
    def preorder_print(self, start, traversal):
        """Helper method - use this to create a 
        recursive print solution."""
        if start.value is None:
            return traversal
        elif start.value is not None:
            traversal = traversal+str(start.value)
            nodeleft = start.left
            noderight = start.right
            if nodeleft is not None:
                traversal = traversal+"-"
                traversal=self.preorder_print(nodeleft,traversal)
                if nodeleft.right is not None:
                    traversal = traversal+"-"
                    traversal = self.preorder_print(nodeleft.right,traversal)
                else:
                    return traversal
            if noderight is not None:
                traversal = traversal+"-"
                traversal = self.preorder_print(noderight,traversal)
            #else:
            #    return traversal
        return traversal
```
